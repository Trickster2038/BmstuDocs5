```python
import pandas as pd

df = pd.DataFrame()
i = 0
with open("access.log") as inf:
     for line in inf:
            tokens = line.split()
            df_buf = pd.DataFrame([tokens])
            df = df.append(df_buf, ignore_index=True)
            i +=1
#             print(df_buf)
            if i>1000:
                break
print(df)
```

```python
names=["ip", "1", "2", "3", "4", "url", "status", "size", "5", "6", "7"]
df3 = pd.read_csv("access.log", sep=" ", usecols= range(11), low_memory=False, header=None, names=names)
```

```python
df3 
```

```python
df4 = pd.concat([df3.iloc[:,0],df3.iloc[:,4:]], axis=1)
df4
```

```python
df4.shape[0]
```

```python
df_urls = pd.DataFrame(df4.iloc[:,2])
df_urls
```

```python
df_urls["cnt"] = 0
df_urls_g = pd.DataFrame(df_urls.groupby(by=["url"]).count())
df_urls_g.sort_values("cnt", ascending=False).head(10)
# df_urls_g.head(10)
```

```python
df5 = df4
# df5["method"] = df5["4"].split()[0]
df5["method"] = 0
df5["method"] = df5.apply (lambda row: row["url"].split()[0], axis=1)
df5
```

```python
df_methods = pd.DataFrame(df5["method"])
df_methods["cnt"] = 0
df_meths = pd.DataFrame(df_methods.groupby(by=["method"]).count())
df_meths
```
