```python
import pandas as pd
```

```python
names=["ip", "1", "2", "3", "4", "url", "status", "size", "5", "6", "7"]
df3 = pd.read_csv("access.log", sep=" ", usecols= range(11), low_memory=False, header=None, names=names)
```

```python
df3 
```

```python
df4 = pd.concat([df3.iloc[:,0],df3.iloc[:,4:]], axis=1)
df4
```

```python
df4.shape[0]
```

```python
df_urls = pd.DataFrame(df4.iloc[:,2])
df_urls
```

```python
df_urls["cnt"] = 0
df_urls2 = df_urls.copy()
df_urls2["url_without_params"] = df_urls.apply (lambda row: row["url"].split()[1], axis=1)
del df_urls2["url"]
df_urls_g = df_urls2.groupby(by=["url_without_params"], as_index=False).count()
df_top_10 = df_urls_g.sort_values("cnt", ascending=False).head(10)
print(df_top_10.to_string(index=False))
# df_urls_g.head(10)
```

```python
df5 = df4
# df5["method"] = df5["4"].split()[0]
df5["method"] = 0
df5["method"] = df5.apply (lambda row: row["url"].split()[0], axis=1)
df5
```

```python
df_methods = pd.DataFrame(df5["method"])
df_methods["cnt"] = 0
df_meths = df_methods.groupby(by=["method"], as_index=False).count()
# print(df_meths)
df_m_normal = df_meths.loc[(df_meths["method"].astype('str').str.len() < 10)]
print(df_m_normal.to_string(index=False))
```

```python
df_m_unnormal = df_meths.loc[(df_meths["method"].astype('str').str.len() >= 10)]
print(df_m_unnormal.to_string(index=False))
```

```python
df_4xx = df4.query("status >= 400 and status < 500").iloc[:,[0,2,3,4]].sort_values("size", ascending=False).head(5)
print(df_4xx.to_string(index=False))
```

```python
df_5xx = df4.query("status >= 500 and status < 600").iloc[:,[0]]
df_5xx["cnt"] = 0
df_5xx_ans = df_5xx.groupby(by=["ip"], as_index=False).count().sort_values("cnt", ascending=False).head(5)
print(df_5xx_ans.to_string(index=False))
```
