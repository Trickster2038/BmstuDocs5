```python
import pandas as pd
```

```python
names=["ip", "1", "2", "3", "4", "url", "status", "size", "5", "6", "7"]
df = pd.read_csv("access.log", sep=" ", usecols= range(11), low_memory=False, header=None, names=names)
df = df.iloc[:,[0,5,6,7,8,9,10]]
```

```python
total_queries = df.shape[0]
total_queries
```

```python
df_methods_1 = df.copy()
df_methods_1["method"] = df_methods_1.apply (lambda row: row["url"].split()[0], axis=1)
df_methods_2 = pd.DataFrame(df_methods_1["method"])
df_methods_2["cnt"] = 0
df_meths = df_methods_2.groupby(by=["method"], as_index=False).count()
df_m_normal = df_meths.loc[(df_meths["method"].astype('str').str.len() < 10)]
df_m_normal
```

```python
df_m_unnormal = df_meths.loc[(df_meths["method"].astype('str').str.len() >= 10)]
df_m_unnormal
```

```python
df_urls = pd.DataFrame(df.iloc[:,1])
df_urls["cnt"] = 0
df_urls2 = df_urls.copy()
df_urls2["url_without_params"] = df_urls.apply (lambda row: row["url"].split()[1], axis=1)
del df_urls2["url"]
df_urls_g = df_urls2.groupby(by=["url_without_params"], as_index=False).count()
df_top_10 = df_urls_g.sort_values("cnt", ascending=False).head(10)
df_top_10
# df_urls
```

```python
df_4xx = df.query("status >= 400 and status < 500").iloc[:,0:4].sort_values("size", ascending=False).head(5)
df_4xx
```

```python
df_5xx = df.query("status >= 500 and status < 600").iloc[:,[0]]
df_5xx["cnt"] = 0
df_5xx_ans = df_5xx.groupby(by=["ip"], as_index=False).count().sort_values("cnt", ascending=False).head(5)
df_5xx_ans
```
